## 미분의 정의
$$  f'(x) = \frac{dy}{dx} = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}  $$

```python
def derivative(f,x):
    h =0.0001
    return (f(x+h) - f(x))/h
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
```

```python
def f(x):
    return 2*x**2+6*x
```

```python
y = f(x)
```

```python
plt.plot(x,y, label="f'(x) = 2x^2 + 6x")
dy_dx = derivative(f, x)
plt.plot(x, dy_dx, label="f'(x) = 4x + 6", linestyle="--")
```

```python
f_sin = np.sin(x)
```

```python
dy_dx_sin = derivative(np.sin, x)
```

```python
plt.plot(x, f_sin, label='sin')
plt.plot(x, dy_dx_sin, label="f'(x)=cos(x)", linestyle="--")
```

```python
f_cos = np.cos(x)
```

```python
dy_dx_cos = derivative(np.cos, x)
```

```python
plt.plot(x, f_cos, label='cos')
plt.plot(x, dy_dx_cos, label="f'(x)=sin(x)", linestyle="--")
```

```python
x = np.linspace(0, 10, 500)

# 주파수 계수들과 선 스타일을 리스트로 정의
factors = [0.5, 0.3, 0.1]
linestyles = ["-", "--", "-."]

# 반복문으로 그래프 그리기
for factor, style in zip(factors, linestyles):
    y = np.sin(factor * np.pi * x)
    plt.plot(x, y, linestyle=style, label=f"sin({factor}πx)")

plt.title("Multiple sin(πx) curves")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
